```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df.drop('Loan_ID', axis=1, inplace=True)
```

```python
# Visualize outliers using boxplots for numerical columns
plt.figure(figsize=(12, 8))
sns.boxplot(data=df)
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
df.isnull().sum()
```

```python
print('Number of people who took loan by gender')
print(df['Gender'].value_counts())
sns.countplot(x='Gender',data = df, palette='Set1')
```

```python
print('Number of people who took loan by married status')
print(df['Married'].value_counts())
sns.countplot(x='Married',data = df, palette='Set1')
```

```python
print('Number of people who took loan by education status')
print(df['Education'].value_counts())
sns.countplot(x='Education',data = df, palette='Set1')
```

```python
print('Number of people who took loan by self employed status')
print(df['Self_Employed'].value_counts())
sns.countplot(x='Self_Employed',data = df, palette='Set1')
```

```python
df['Total_Income'] = df['ApplicantIncome'] + df['CoapplicantIncome']
df.head()
```

```python
df['ApplicantIncomelog'] = np.log(df['ApplicantIncome'] + 1)
sns.distplot(df['ApplicantIncomelog'])
```

```python
df['LoanAmountlog'] = np.log(df['LoanAmount'] + 1)
sns.distplot(df['LoanAmountlog'])
```

```python
df['Loan_Amount_Term_log'] = np.log(df['Loan_Amount_Term'] + 1)
sns.distplot(df['Loan_Amount_Term_log'])
```

```python
df['Total_Income_log'] = np.log(df['Total_Income'] + 1)
sns.distplot(df['Total_Income_log'])
```

```python
cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Total_Income']
df = df.drop(columns = cols, axis = 1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ['Gender','Married','Education','Dependents','Self_Employed','Property_Area','Loan_Status']
le =  LabelEncoder()
for col in cols:
  df[col] =  le.fit_transform(df[col])
```

```python
df.head()
```

```python
df.dtypes
```

```python
X = df.drop(columns = ['Loan_Status'],axis = 1)
y = df['Loan_Status']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25,random_state = 42)
```

```python
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train,y_train)
```

```python
y_pred = log_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Logistic Regression Accuracy:", accuracy)
print("Logistic Regression Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Cross Validation score of Logistic Regresion: ", cross_val_score(log_model, X_train, y_train, cv=5).mean())
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train,y_train)
```

```python
y_pred_tree = tree_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred_tree)
```

```python
print("Accuracy score of Decision Tree: ", accuracy)
print("Confusion Matrix of Decision Tree: \n", confusion_matrix(y_test, y_pred_tree))
print("Cross Validation score of Decision Tree: ", cross_val_score(tree_model, X, y, cv=5).mean())
```

```python
forest_model = RandomForestClassifier()
forest_model.fit(X_train,y_train)
```

```python
y_pred_forest = forest_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred_forest)
```

```python
print("Accuracy score of Random Forest: ", accuracy)
print("Confusion Matrix of Random Forest: \n", confusion_matrix(y_test, y_pred_forest))
print("Cross Validation score of Random Forest: ", cross_val_score(forest_model, X, y, cv=5).mean())
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train,y_train)
```

```python
y_pred_knn = knn_model.predict(X_test)
accuracy = accuracy_score(y_test,y_pred_knn)
```

```python
print("Accuracy score of KNeighbors: ", accuracy)
print("Confusion Matrix of KNeighbors: \n", confusion_matrix(y_test, y_pred_forest))
print("Cross Validation score of KNeighbors: ", cross_val_score(forest_model, X, y, cv=5).mean())
```

```python
from sklearn.metrics import classification_report

def generate_classification_report(model_name,y_test,y_pred):
  report = classification_report(y_test,y_pred)
  print(f"Classification Report For {model_name}:\n{report}\n")

generate_classification_report(log_model,y_test,y_pred)
generate_classification_report(tree_model,y_test,y_pred_tree)
generate_classification_report(forest_model,y_test,y_pred_forest)
generate_classification_report(knn_model,y_test,y_pred_knn)
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
oversample = RandomOverSampler(random_state=42)
X_resampled, y_resampled = oversample.fit_resample(X,y)

df_resampled = pd.concat([pd.DataFrame(X_resampled,columns=X.columns),pd.Series(y_resampled,name="Loan_status")],axis=1)
```

```python
X_resampled
```

```python
y_resampled
```

```python
y_resampled.value_counts()
```

```python
X_resampled_train, X_resampled_test, y_resampled_train, y_resampled_test = train_test_split(X_resampled,y_resampled,test_size = 0.25,random_state=42)
```

```python
## Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_resampled_train,y_resampled_train)
y_pred = log_model.predict(X_resampled_test)
accuracy = accuracy_score(y_resampled_test,y_pred)
accuracy*100
```

```python
## Decision Tree Classifier

tree_model = DecisionTreeClassifier()
tree_model.fit(X_resampled_train,y_resampled_train)
y_pred_tree = tree_model.predict(X_resampled_test)
accuracy = accuracy_score(y_resampled_test,y_pred_tree)
print("Accuracy score of Decision Tree: ", accuracy*100)
```

```python
## Random Forest Classifier
forest_model = RandomForestClassifier()
forest_model.fit(X_resampled_train,y_resampled_train)
y_pred_forest = forest_model.predict(X_resampled_test)
accuracy = accuracy_score(y_resampled_test,y_pred_forest)
print("Accuracy score of Random Forest: ", accuracy*100)
```

```python
#KNearestNeighbors model
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_resampled_train,y_resampled_train)
y_pred_knn = knn_model.predict(X_resampled_test)
accuracy = accuracy_score(y_resampled_test,y_pred_knn)
print("Accuracy score of KNeighbors: ", accuracy*100)
```

```python
from sklearn.metrics import classification_report

def generate_classification_report(model_name,y_test,y_pred):
  report = classification_report(y_test,y_pred)
  print(f"Classification Report For {model_name}:\n{report}\n")

generate_classification_report(log_model,y_resampled_test,y_pred)
generate_classification_report(tree_model,y_resampled_test,y_pred_tree)
generate_classification_report(forest_model,y_resampled_test,y_pred_forest)
generate_classification_report(knn_model,y_resampled_test,y_pred_knn)
```
